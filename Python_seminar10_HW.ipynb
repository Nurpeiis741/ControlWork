```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import random
lst = ['robot'] * random.randint(5, 15)
lst += ['human'] * random.randint(5, 15)
lst += ['alien'] * random.randint(5, 15)
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
data.loc[data['whoAmI'] == 'robot', 'one_hot'] = 0
data.loc[data['whoAmI'] == 'human', 'one_hot'] = 1
data.loc[data['whoAmI'] == 'alien', 'one_hot'] = 2
data['one_hot']
```

```python
sns.histplot(data = data, x='whoAmI')
```

```python
pd.get_dummies(data['whoAmI'])
```

```python
data.loc[data['whoAmI'] == 'robot', 'robot'] = True
data.loc[data['whoAmI'] != 'robot', 'robot'] = False
data.loc[data['whoAmI'] == 'human', 'human'] = True
data.loc[data['whoAmI'] != 'human', 'human'] = False
data[['robot', 'human']]
```

```python
pd.get_dummies(data['whoAmI'])
```

```python
sns.histplot(data=data, x='robot', hue='human')
```
